**6. Custom Training**

```python
from ultralytics import YOLO
import os

# Path of the datasets
ROOT_DIR = "/content/MY-project-3"

# Build a model from pretrained weight
model = YOLO("yolov9e-seg.pt")

# Train the model
results = model.train(data=os.path.join(ROOT_DIR, "data.yaml"), epochs=1000, imgsz=640, single_cls=True)
```
